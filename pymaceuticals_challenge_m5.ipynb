## Quartiles, Outliers, and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

# study results, drop g989, set stat_analysis_4

stat_analysis_4 = study_results[study_results['Mouse ID'] != 'g989']

# keep columns that are relevant to me at this time

stat_analysis_4 = stat_analysis_4[['Mouse ID', 'Drug Regimen', 'Weight (g)', 'Timepoint', 'Tumor Volume (mm3)', ]]

# Keep only the row with the greatest 'Timepoint' value for each 'Mouse ID'
# DID IT 3 TIMES IN SLGIHTLY DIFFERENT WAYS FIRST, VERIFIED THAT THIS DOES CONTAIN THE HIGHEST VALUES FOR EACH MOUSE ID, 
# IMPLYING THAT MICE DIED BEFORE STUDY END. ISOLATED INDIVIDUAL IDS WITH QUESTIONABLE NUMBERS IN TIMEPOINT

stat_analysis_4 = stat_analysis_4.loc[stat_analysis_4.groupby('Mouse ID')['Timepoint'].idxmax()]

stat_analysis_4

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
# DON'T REALLY UNDERSTAND THE PURPOSE OF THIS SO I'M MOVING ON. THAT HAS ALREADY BEEN ACCOMPLISHED AS FAR AS I KNOW.
```

```python
# isolate the particular drugs that we care about. the above is just showing the steps so that I know I've done my work correctly.

drug_names = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
stat_analysis_4 = stat_analysis_4[stat_analysis_4['Drug Regimen'].isin(drug_names)]

stat_analysis_4
```

```python
# Treatment names list, empty list
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
tumor_vol = []

# For Loop
for drug_name in treatments:
    # Locate tumor volumes for each drug
    tumor_volumes = stat_analysis_4.loc[stat_analysis_4['Drug Regimen'] == drug_name]['Tumor Volume (mm3)']
    
    # Add tumor volumes to list
    tumor_vol.append(tumor_volumes)

    # Quartiles
    quartiles = tumor_volumes.quantile([0.25, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq

    # Outliers
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    outliers = tumor_volumes.loc[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
    
    # Print
    print(f"Drug Name: {drug_name}")
    print(f"Lower Quartile: {lowerq}")
    print(f"Upper Quartile: {upperq}")
    print(f"IQR: {iqr}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Potential Outliers: {outliers}")
    print(" ")
```

```python
# Box Plot
plt.boxplot(tumor_vol, labels=treatments, flierprops=dict(marker='o', markerfacecolor='red', markersize=8))
plt.xlabel('Drug Regimen')
plt.ylabel('Final Tumor Volume (mm3)')
plt.title('Final Tumor Volume Distribution')
plt.show()
```

```python
# Calculate the average tumor volume for mice in Capomulin regimen
capomulin_data = study_results[study_results['Drug Regimen'] == 'Capomulin']
avg_tumor_volume = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()

# Merged average tumor volume with the mouse weight
merged_vol_weight = pd.merge(avg_tumor_volume, capomulin_data[['Mouse ID', 'Weight (g)']], on='Mouse ID', how='left').drop_duplicates()

# Scatter plot
plt.scatter(merged_vol_weight['Weight (g)'], merged_vol_weight['Tumor Volume (mm3)'], marker='o', facecolors='green', edgecolors='black')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')
plt.grid(False)
plt.show()
```

## Correlation and Regression

```python
# Calculate and print correlation coefficient
correlation = merged_vol_weight['Weight (g)'].corr(merged_vol_weight['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is: {correlation:.2f}")

# Linear Regression
slope, intercept, rvalue, pvalue, stderr = linregress(merged_vol_weight['Weight (g)'], merged_vol_weight['Tumor Volume (mm3)'])

# Regression Values
regress_values = merged_vol_weight['Weight (g)'] * slope + intercept

# Scatter Plot
plt.scatter(merged_vol_weight['Weight (g)'], merged_vol_weight['Tumor Volume (mm3)'], marker='o', facecolors='blue', edgecolors='black')

# Regression Line
plt.plot(merged_vol_weight['Weight (g)'], regress_values, color='red')

plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Mouse Weight vs. Average Tumor Volume (Capomulin)')

plt.show()

```
